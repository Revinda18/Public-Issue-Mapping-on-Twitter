# DATA CLEANING

```python
import pandas as pd
import re

df = pd.read_csv("/content/dataset_kemenkeu_indo.csv")

# Fungsi untuk membersihkan teks
def clean_text(text):
    if pd.isnull(text):
        return ""

    # Menghapus karakter non-ASCII
    text = text.encode("ascii", "ignore").decode()

    # Menghapus emoji
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticon
        "\U0001F300-\U0001F5FF"  # simbol & pictographs
        "\U0001F680-\U0001F6FF"  # simbol transportasi & peta
        "\U0001F1E0-\U0001F1FF"  # bendera
        "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Menghapus tab, newline, dan karakter escape lainnya
    text = re.sub(r'[\t\n\r\\]', ' ', text)

    # Menghapus kata "RT"
    text = re.sub(r'\bRT\b', '', text)

    # Menghapus mention (@username)
    text = re.sub(r'@\w+', '', text)

    # Menghapus URL yang tidak lengkap (tidak diawali http atau www)
    text = re.sub(r'\b(?!http|www)\w+\.\w+\S*', '', text)

    # Menghapus tanda hubung dengan spasi
    text = re.sub(r'\s*-\s*', ' ', text)

    # Mengubah semua teks menjadi huruf kecil
    text = text.lower()

    # Menghapus tanda baca kecuali underscore (_)
    text = re.sub(r'[^\w\s_]', '', text)

    # Mengganti tanda hubung dengan spasi
    text = text.replace('-', ' ')

    # Menghapus kata "user"
    text = re.sub(r'\buser\b', '', text)

    # Menghapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Terapkan fungsi cleaning ke kolom 'full_text'
df['cleaned_text'] = df['full_text'].apply(clean_text)

# Hapus duplikat berdasarkan kolom cleaned_text
df_cleaned = df.drop_duplicates(subset='cleaned_text')

# Tampilkan 10 baris pertama dari teks asli dan hasil pembersihannya
print(df_cleaned[['full_text', 'cleaned_text']].head(5))
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.metrics.pairwise import cosine_similarity
from wordcloud import WordCloud
from itertools import combinations

# Load dataset
df = pd.read_csv("/content/dataset_kemenkeu_cleaned.csv")
df.dropna(subset=["full_text"], inplace=True)
df.drop_duplicates(subset=["full_text"], inplace=True)

# Normalisasi teks
def normalize_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"@\w+|\#", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text.strip()

df["normalisasi_text"] = df["full_text"].apply(normalize_text)

# Vectorizer
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2)
doc_term_matrix = vectorizer.fit_transform(df["normalisasi_text"])

# Fungsi menghitung coherence score tanpa gensim
def compute_coherence_score(top_words_list, doc_term_matrix, vectorizer):
    feature_names = vectorizer.get_feature_names_out()
    word_idx = {word: i for i, word in enumerate(feature_names)}

    scores = []
    for topic_words in top_words_list:
        combis = list(combinations(topic_words, 2))
        topic_score = 0
        valid_combis = 0

        for w1, w2 in combis:
            if w1 in word_idx and w2 in word_idx:
                idx1 = word_idx[w1]
                idx2 = word_idx[w2]

                vec1 = doc_term_matrix[:, idx1].toarray()
                vec2 = doc_term_matrix[:, idx2].toarray()

                sim = cosine_similarity(vec1, vec2)[0][0]
                topic_score += sim
                valid_combis += 1

        if valid_combis > 0:
            scores.append(topic_score / valid_combis)

    return np.mean(scores) if scores else 0

# Coba berbagai jumlah topik
topic_range = range(2, 11)
perplexities = []
coherence_scores = []
models = []

for n_topics in topic_range:
    lda = LatentDirichletAllocation(n_components=n_topics, max_iter=10,
                                    learning_method='batch',
                                    random_state=42)
    lda.fit(doc_term_matrix)
    models.append(lda)

    # Hitung perplexity
    perplexities.append(lda.perplexity(doc_term_matrix))

    # Ambil top words untuk setiap topik
    words = vectorizer.get_feature_names_out()
    top_words_list = [
        [words[i] for i in topic.argsort()[:-11:-1]]
        for topic in lda.components_
    ]

    # Hitung coherence score
    coherence = compute_coherence_score(top_words_list, doc_term_matrix, vectorizer)
    coherence_scores.append(coherence)

# Plot perplexity dan coherence
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(topic_range, perplexities, marker='o', color='red')
plt.title("Perplexity untuk Berbagai Jumlah Topik")
plt.xlabel("Jumlah Topik")
plt.ylabel("Perplexity (semakin kecil semakin baik)")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(topic_range, coherence_scores, marker='o', color='green')
plt.title("Coherence Score untuk Berbagai Jumlah Topik")
plt.xlabel("Jumlah Topik")
plt.ylabel("Coherence Score (semakin besar semakin baik)")
plt.grid(True)

plt.tight_layout()
plt.show()

# Rangkuman hasil evaluasi
print("="*60)
print("Rangkuman Hasil Evaluasi Jumlah Topik")
for i, n in enumerate(topic_range):
    print(f"Topik: {n} | Perplexity: {perplexities[i]:.2f} | Coherence: {coherence_scores[i]:.4f}")
print("="*60)

# Pilih model terbaik berdasarkan coherence
best_idx = np.argmax(coherence_scores)
optimal_topics = topic_range[best_idx]
best_lda_model = models[best_idx]

print(f"Jumlah topik optimal berdasarkan coherence: {optimal_topics}")
print(f"Skor coherence tertinggi: {coherence_scores[best_idx]:.4f}")
print("="*60)

# Tampilkan topik-topik
words = vectorizer.get_feature_names_out()
for idx, topic in enumerate(best_lda_model.components_):
    print(f"Topik {idx}: ", [words[i] for i in topic.argsort()[:-11:-1]])

# Wordcloud per topik
for idx, topic in enumerate(best_lda_model.components_):
    plt.figure()
    word_freq = {words[i]: topic[i] for i in topic.argsort()[:-21:-1]}
    wordcloud = WordCloud(background_color='white').generate_from_frequencies(word_freq)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f"Wordcloud untuk Topik {idx}")
    plt.show()

# Distribusi dokumen-topik
doc_topic_distrib = best_lda_model.transform(doc_term_matrix)
topic_df = pd.DataFrame(doc_topic_distrib)

# Korelasi antar topik
correlation_matrix = topic_df.corr(method='pearson')

# Heatmap korelasi
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelasi Pearson antar Topik")
plt.show()

# Cari korelasi tertinggi selain diagonal
corr_values = correlation_matrix.values.copy()
np.fill_diagonal(corr_values, np.nan)
max_corr_value = np.nanmax(corr_values)
max_corr_indices = np.unravel_index(np.nanargmax(corr_values), corr_values.shape)

print("="*60)
print(f"Pasangan topik dengan korelasi tertinggi:")
print(f"Topik {max_corr_indices[0]} dan Topik {max_corr_indices[1]} dengan nilai korelasi: {max_corr_value:.4f}")
print("="*60)

words = vectorizer.get_feature_names_out()
n_top_words = 10  # Jumlah kata per topik yang ingin ditampilkan

for idx, topic in enumerate(best_lda_model.components_):
    print(f"Topik {idx} Kata:", end=" ")
    top_words = topic.argsort()[:-n_top_words - 1:-1]
    top_words_str = " + ".join([f"{topic[i]:.3f}*\"{words[i]}\"" for i in top_words])
    print(top_words_str)

# Definisikan stopwords tambahan (selain dari vectorizer)
extra_stopwords = set([
    "dan", "yang", "di", "ini", "untuk", "dengan", "dalam", "pada",
    "adalah", "itu", "kami", "juga", "oleh", "akan", "yg", "sebagai"
])

words = vectorizer.get_feature_names_out()
top_keywords_per_topic = []

print("="*60)
print("KATA KUNCI BERSIH PER TOPIK (TANPA STOPWORDS)")
for idx, topic in enumerate(best_lda_model.components_):
    top_indices = topic.argsort()[:-21:-1]
    raw_keywords = [words[i] for i in top_indices]
    filtered_keywords = [word for word in raw_keywords if word not in extra_stopwords]
    top_keywords_per_topic.append(filtered_keywords)
    print(f"Topik {idx}: {', '.join(filtered_keywords)}")
```

```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.metrics.pairwise import cosine_similarity
from wordcloud import WordCloud
from itertools import combinations

# Load dan persiapan data
df = pd.read_csv("/content/dataset_kemenkeu_cleaned.csv")
df.dropna(subset=["full_text"], inplace=True)
df.drop_duplicates(subset=["full_text"], inplace=True)

# Normalisasi teks
def normalize_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"@\w+|\#", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text.strip()

df["normalisasi_text"] = df["full_text"].apply(normalize_text)

# Daftar stopword Bahasa Indonesia
indonesian_stopwords = """
ada adakah adalah adakala adapun ad interim aku anda andaikata andaikan
antaranya apabila adalah aku aku anda andaikata
atau oleh karena itu oleh sebab itu oleh karenanya sebab oleh sebab
dan di dalam di atas di bawah dia mereka saya kita kamu kamu sendiri
untuk dengan dari sampai ke tanpa tentang serta kalau kalaupun jika
tetapi tapi bahwa bahwa juga sudah belum maka tidak bukan malah
bahwa akan supaya agar demi demi demi supaya guna guna guna agar supaya
dari dari sejak sejak sejak oleh oleh oleh
""".split()

# --- Pengecekan dan pembuatan kolom 'stopword' ---
if 'stopword' not in df.columns:
    def filter_stopwords(text):
        words = text.split()
        filtered = [word for word in words if word in indonesian_stopwords]
        return " ".join(filtered)

    df['stopword'] = df['normalisasi_text'].apply(filter_stopwords)

# Selanjutnya proses vektorisasi dan LDA tetap sama
vectorizer = CountVectorizer(stop_words=indonesian_stopwords, max_df=0.95, min_df=2)
doc_term_matrix = vectorizer.fit_transform(df["normalisasi_text"])
feature_names = vectorizer.get_feature_names_out()

def compute_coherence_score(top_words_list, doc_term_matrix, vectorizer):
    word_idx = {word: i for i, word in enumerate(vectorizer.get_feature_names_out())}
    scores = []
    for topic_words in top_words_list:
        combis = list(combinations(topic_words, 2))
        topic_score = 0
        valid_combis = 0
        for w1, w2 in combis:
            if w1 in word_idx and w2 in word_idx:
                idx1, idx2 = word_idx[w1], word_idx[w2]
                vec1 = doc_term_matrix[:, idx1].toarray()
                vec2 = doc_term_matrix[:, idx2].toarray()
                sim = cosine_similarity(vec1, vec2)[0][0]
                topic_score += sim
                valid_combis += 1
        if valid_combis > 0:
            scores.append(topic_score / valid_combis)
    return np.mean(scores) if scores else 0

topic_range = range(2, 11)
perplexities = []
coherences = []
models = []

for n_topics in topic_range:
    lda = LatentDirichletAllocation(n_components=n_topics, max_iter=10,
                                    learning_method='batch',
                                    random_state=42)
    lda.fit(doc_term_matrix)
    models.append(lda)

    perplexities.append(lda.perplexity(doc_term_matrix))
    words = vectorizer.get_feature_names_out()
    top_words_list = [[words[i] for i in topic.argsort()[:-11:-1]] for topic in lda.components_]
    coherence = compute_coherence_score(top_words_list, doc_term_matrix, vectorizer)
    coherences.append(coherence)

plt.figure(figsize=(10,5))
plt.plot(topic_range, perplexities, label="Perplexity", marker='o')
plt.plot(topic_range, coherences, label="Coherence (manual)", marker='s')
plt.xlabel("Jumlah Topik")
plt.ylabel("Nilai")
plt.title("Perbandingan Perplexity dan Coherence dengan Stopword Bahasa Indonesia")
plt.legend()
plt.grid(True)
plt.show()

optimal_index = np.argmax(coherences)
optimal_topic = topic_range[optimal_index]
final_lda = models[optimal_index]
print(f"Jumlah topik optimal (coherence tertinggi): {optimal_topic}")
print(f"Coherence terbaik: {coherences[optimal_index]:.4f}")

doc_topic = final_lda.transform(doc_term_matrix)
df['topik_dominan'] = doc_topic.argmax(axis=1)

plt.figure(figsize=(8,5))
sns.countplot(x='topik_dominan', data=df, palette='viridis')
plt.title('Distribusi Topik Dominan pada Dokumen')
plt.xlabel('Topik Dominan')
plt.ylabel('Jumlah Dokumen')
plt.show()

topic_df = pd.DataFrame(doc_topic, columns=[f'Topik {i}' for i in range(optimal_topic)])
corr_matrix = topic_df.corr()

plt.figure(figsize=(10,7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Pearson Antar Topik')
plt.show()

def print_topics_with_weights(lda_model, feature_names, n_top_words=10):
    for topic_idx, topic in enumerate(lda_model.components_):
        topic_weights = topic / topic.sum()
        top_indices = topic_weights.argsort()[:-n_top_words - 1:-1]
        topic_words = [f'{topic_weights[i]:.3f}*"{feature_names[i]}"' for i in top_indices]
        topic_str = " + ".join(topic_words)
        print(f"Topik {topic_idx} Kata: {topic_str}")

print_topics_with_weights(final_lda, feature_names, 10)

fig, axes = plt.subplots(1, optimal_topic, figsize=(5 * optimal_topic, 5), constrained_layout=True)
if optimal_topic == 1:
    axes = [axes]

for topic_idx, topic in enumerate(final_lda.components_):
    freq = {feature_names[i]: topic[i] for i in topic.argsort()[:-30 - 1:-1]}
    wordcloud = WordCloud(width=400, height=300, background_color='white').generate_from_frequencies(freq)
    ax = axes[topic_idx]
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.axis('off')
    ax.set_title(f"Topik {topic_idx}")

plt.show()


def print_topics_with_weights_and_keywords(lda_model, feature_names, stopwords, n_top_words=10):
    print("=== Topik dengan bobot kata ===")
    for topic_idx, topic in enumerate(lda_model.components_):
        topic_weights = topic / topic.sum()
        top_indices = topic_weights.argsort()[:-n_top_words - 1:-1]
        topic_words_weights = [f'{topic_weights[i]*1000:.3f}*"{feature_names[i]}"' for i in top_indices]
        topic_str = " + ".join(topic_words_weights)
        print(f"Topik {topic_idx} Kata: {topic_str}")

    print("\n=== KATA KUNCI BERSIH PER TOPIK (TANPA STOPWORDS) ===")
    for topic_idx, topic in enumerate(lda_model.components_):
        topic_weights = topic / topic.sum()
        top_indices = topic_weights.argsort()[:-30 - 1:-1]  # ambil 30 kata teratas
        top_words = [feature_names[i] for i in top_indices if feature_names[i] not in stopwords]
        print(f"Topik {topic_idx}: {', '.join(top_words)}")

# Panggil fungsi ini dengan parameter yang sesuai
print_topics_with_weights_and_keywords(final_lda, feature_names, indonesian_stopwords, n_top_words=10)
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.metrics.pairwise import cosine_similarity
from wordcloud import WordCloud
from itertools import combinations

# Import stemmer dari Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Load dataset
df = pd.read_csv("/content/dataset_kemenkeu_cleaned.csv")
df.dropna(subset=["full_text"], inplace=True)
df.drop_duplicates(subset=["full_text"], inplace=True)

# Normalisasi teks
def normalize_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"@\w+|\#", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text.strip()

df["normalisasi_text"] = df["full_text"].apply(normalize_text)

# Inisialisasi stemmer Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi stemming menggunakan Sastrawi
def stem_text(text):
    return stemmer.stem(text)

# Terapkan stemming ke teks yang sudah dinormalisasi
df["stemmed_text"] = df["normalisasi_text"].apply(stem_text)

# Vectorizer tanpa stopword removal
vectorizer = CountVectorizer(max_df=0.95, min_df=2)
doc_term_matrix = vectorizer.fit_transform(df["stemmed_text"])

# Fungsi menghitung coherence score tanpa gensim
def compute_coherence_score(top_words_list, doc_term_matrix, vectorizer):
    feature_names = vectorizer.get_feature_names_out()
    word_idx = {word: i for i, word in enumerate(feature_names)}

    scores = []
    for topic_words in top_words_list:
        combis = list(combinations(topic_words, 2))
        topic_score = 0
        valid_combis = 0

        for w1, w2 in combis:
            if w1 in word_idx and w2 in word_idx:
                idx1 = word_idx[w1]
                idx2 = word_idx[w2]

                vec1 = doc_term_matrix[:, idx1].toarray()
                vec2 = doc_term_matrix[:, idx2].toarray()

                sim = cosine_similarity(vec1, vec2)[0][0]
                topic_score += sim
                valid_combis += 1

        if valid_combis > 0:
            scores.append(topic_score / valid_combis)

    return np.mean(scores) if scores else 0

# Eksperimen jumlah topik LDA
topic_range = range(2, 11)
perplexities = []
coherence_scores = []
models = []

for n_topics in topic_range:
    lda = LatentDirichletAllocation(n_components=n_topics, max_iter=10,
                                    learning_method='batch',
                                    random_state=42)
    lda.fit(doc_term_matrix)
    models.append(lda)

    perplexities.append(lda.perplexity(doc_term_matrix))

    words = vectorizer.get_feature_names_out()
    top_words_list = [
        [words[i] for i in topic.argsort()[:-11:-1]]
        for topic in lda.components_
    ]

    coherence = compute_coherence_score(top_words_list, doc_term_matrix, vectorizer)
    coherence_scores.append(coherence)

# Visualisasi hasil
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(topic_range, perplexities, marker='o', color='red')
plt.title("Perplexity untuk Berbagai Jumlah Topik")
plt.xlabel("Jumlah Topik")
plt.ylabel("Perplexity (semakin kecil semakin baik)")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(topic_range, coherence_scores, marker='o', color='green')
plt.title("Coherence Score untuk Berbagai Jumlah Topik")
plt.xlabel("Jumlah Topik")
plt.ylabel("Coherence Score (semakin besar semakin baik)")
plt.grid(True)

plt.tight_layout()
plt.show()

# Cetak rangkuman
print("="*60)
print("Rangkuman Hasil Evaluasi Jumlah Topik")
for i, n in enumerate(topic_range):
    print(f"Topik: {n} | Perplexity: {perplexities[i]:.2f} | Coherence: {coherence_scores[i]:.4f}")
print("="*60)

# Pilih model terbaik
best_idx = np.argmax(coherence_scores)
optimal_topics = topic_range[best_idx]
best_lda_model = models[best_idx]

print(f"Jumlah topik optimal berdasarkan coherence: {optimal_topics}")
print(f"Skor coherence tertinggi: {coherence_scores[best_idx]:.4f}")
print("="*60)

# Tampilkan topik
words = vectorizer.get_feature_names_out()
for idx, topic in enumerate(best_lda_model.components_):
    print(f"Topik {idx}: ", [words[i] for i in topic.argsort()[:-11:-1]])

# Wordcloud per topik
for idx, topic in enumerate(best_lda_model.components_):
    plt.figure()
    word_freq = {words[i]: topic[i] for i in topic.argsort()[:-21:-1]}
    wordcloud = WordCloud(background_color='white').generate_from_frequencies(word_freq)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f"Wordcloud untuk Topik {idx}")
    plt.show()

# Distribusi dokumen-topik
doc_topic_distrib = best_lda_model.transform(doc_term_matrix)
topic_df = pd.DataFrame(doc_topic_distrib)

# Korelasi antar topik
correlation_matrix = topic_df.corr(method='pearson')

# Heatmap korelasi
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelasi Pearson antar Topik")
plt.show()

# Korelasi tertinggi selain diagonal
corr_values = correlation_matrix.values.copy()
np.fill_diagonal(corr_values, np.nan)
max_corr_value = np.nanmax(corr_values)
max_corr_indices = np.unravel_index(np.nanargmax(corr_values), corr_values.shape)

print("="*60)
print(f"Pasangan topik dengan korelasi tertinggi:")
print(f"Topik {max_corr_indices[0]} dan Topik {max_corr_indices[1]} dengan nilai korelasi: {max_corr_value:.4f}")
print("="*60)

words = vectorizer.get_feature_names_out()
n_top_words = 10  # Jumlah kata per topik yang ingin ditampilkan

for idx, topic in enumerate(best_lda_model.components_):
    print(f"Topik {idx} Kata:", end=" ")
    top_words = topic.argsort()[:-n_top_words - 1:-1]
    top_words_str = " + ".join([f"{topic[i]:.3f}*\"{words[i]}\"" for i in top_words])
    print(top_words_str)

n_top_words = 10
print("="*60)
for idx, topic in enumerate(best_lda_model.components_):
    print(f"Topik {idx} Kata:", end=" ")
    top_words = topic.argsort()[:-n_top_words - 1:-1]
    top_words_str = " + ".join([f"{topic[i]*1000:.3f}*\"{words[i]}\"" for i in top_words])
    print(top_words_str)
print("="*60)

```

## PROCESS D

(NORMALISASI + TOKENISASI + STOPWORD REMOVAL + STEMMING SASTRAWI)

```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.metrics.pairwise import cosine_similarity
from wordcloud import WordCloud
from itertools import combinations
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Load dan persiapan data
df = pd.read_csv("/content/dataset_kemenkeu_cleaned.csv")
df.dropna(subset=["full_text"], inplace=True)
df.drop_duplicates(subset=["full_text"], inplace=True)

# 1. Normalisasi teks
def normalize_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"@\w+|\#", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text.strip()

df["normalisasi_text"] = df["full_text"].apply(normalize_text)

# 2. Tokenisasi (pisah kalimat menjadi kata)
df["tokenized_text"] = df["normalisasi_text"].apply(lambda x: x.split())

# 3. Stopword removal dengan daftar stopword bahasa Indonesia
indonesian_stopwords = set("""
ada adakah adalah adakala adapun ad interim aku anda andaikata andaikan
antaranya apabila adalah aku aku anda anda andaikata
atau oleh karena itu oleh sebab itu oleh karenanya sebab oleh sebab
dan di dalam di atas di bawah dia mereka saya kita kamu kamu sendiri
untuk dengan dari sampai ke tanpa tentang serta kalau kalaupun jika
tetapi tapi bahwa bahwa juga sudah belum maka tidak bukan malah
bahwa akan supaya agar demi demi demi supaya guna guna guna agar supaya
dari dari sejak sejak sejak oleh oleh oleh
""".split())

def remove_stopwords(tokens):
    return [word for word in tokens if word not in indonesian_stopwords]

df["stopword_removed"] = df["tokenized_text"].apply(remove_stopwords)

# 4. Stemming dengan Sastrawi (pada list kata setelah stopword removal)
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]

df["stemmed_tokens"] = df["stopword_removed"].apply(stem_tokens)

# 5. Gabungkan kembali hasil stemming menjadi string untuk vectorizer
df["stemmed_text"] = df["stemmed_tokens"].apply(lambda x: " ".join(x))

# Berikutnya gunakan df["stemmed_text"] untuk vektorisasi dan pemodelan LDA seperti sebelumnya
vectorizer = CountVectorizer(max_df=0.95, min_df=2)
doc_term_matrix = vectorizer.fit_transform(df["stemmed_text"])
feature_names = vectorizer.get_feature_names_out()

# Fungsi hitung coherence manual (tanpa gensim)
def compute_coherence_score(top_words_list, doc_term_matrix, vectorizer):
    word_idx = {word: i for i, word in enumerate(vectorizer.get_feature_names_out())}
    scores = []
    for topic_words in top_words_list:
        combis = list(combinations(topic_words, 2))
        topic_score = 0
        valid_combis = 0
        for w1, w2 in combis:
            if w1 in word_idx and w2 in word_idx:
                idx1, idx2 = word_idx[w1], word_idx[w2]
                vec1 = doc_term_matrix[:, idx1].toarray()
                vec2 = doc_term_matrix[:, idx2].toarray()
                sim = cosine_similarity(vec1, vec2)[0][0]
                topic_score += sim
                valid_combis += 1
        if valid_combis > 0:
            scores.append(topic_score / valid_combis)
    return np.mean(scores) if scores else 0

# Cari jumlah topik terbaik
topic_range = range(2, 11)
perplexities = []
coherences = []
models = []

for n_topics in topic_range:
    lda = LatentDirichletAllocation(n_components=n_topics, max_iter=10,
                                    learning_method='batch',
                                    random_state=42)
    lda.fit(doc_term_matrix)
    models.append(lda)

    perplexities.append(lda.perplexity(doc_term_matrix))
    words = vectorizer.get_feature_names_out()
    top_words_list = [[words[i] for i in topic.argsort()[:-11:-1]] for topic in lda.components_]
    coherence = compute_coherence_score(top_words_list, doc_term_matrix, vectorizer)
    coherences.append(coherence)

# Plot Perplexity dan Coherence
plt.figure(figsize=(10,5))
plt.plot(topic_range, perplexities, label="Perplexity", marker='o')
plt.plot(topic_range, coherences, label="Coherence (manual)", marker='s')
plt.xlabel("Jumlah Topik")
plt.ylabel("Nilai")
plt.title("Perbandingan Perplexity dan Coherence dengan Stopword Bahasa Indonesia dan Stemming")
plt.legend()
plt.grid(True)
plt.show()

# Pilih model dengan coherence tertinggi
optimal_index = np.argmax(coherences)
optimal_topic = topic_range[optimal_index]
final_lda = models[optimal_index]
print(f"Jumlah topik optimal (coherence tertinggi): {optimal_topic}")
print(f"Coherence terbaik: {coherences[optimal_index]:.4f}")

# Prediksi distribusi topik dokumen
doc_topic = final_lda.transform(doc_term_matrix)
df['topik_dominan'] = doc_topic.argmax(axis=1)

# Visualisasi distribusi topik dominan (Bar plot)
plt.figure(figsize=(8,5))
sns.countplot(x='topik_dominan', data=df, palette='viridis')
plt.title('Distribusi Topik Dominan pada Dokumen')
plt.xlabel('Topik Dominan')
plt.ylabel('Jumlah Dokumen')
plt.show()

# Heatmap korelasi antar topik berdasarkan distribusi dokumen
topic_df = pd.DataFrame(doc_topic, columns=[f'Topik {i}' for i in range(optimal_topic)])
corr_matrix = topic_df.corr()

plt.figure(figsize=(10,7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Pearson Antar Topik')
plt.show()

# Fungsi cetak topik dengan bobot kata (normalisasi bobot tiap topik)
def print_topics_with_weights(lda_model, feature_names, n_top_words=10):
    for topic_idx, topic in enumerate(lda_model.components_):
        topic_weights = topic / topic.sum()
        top_indices = topic_weights.argsort()[:-n_top_words - 1:-1]
        topic_words = [f'{topic_weights[i]:.3f}*"{feature_names[i]}"' for i in top_indices]
        topic_str = " + ".join(topic_words)
        print(f"Topik {topic_idx} Kata: {topic_str}")

print_topics_with_weights(final_lda, feature_names, 10)

# Visualisasi WordCloud tiap topik dalam satu figure
fig, axes = plt.subplots(1, optimal_topic, figsize=(5 * optimal_topic, 5), constrained_layout=True)
if optimal_topic == 1:
    axes = [axes]

for topic_idx, topic in enumerate(final_lda.components_):
    freq = {feature_names[i]: topic[i] for i in topic.argsort()[:-30 - 1:-1]}
    wordcloud = WordCloud(width=400, height=300, background_color='white').generate_from_frequencies(freq)
    ax = axes[topic_idx]
    ax.imshow(wordcloud, interpolation='bilinear')
    ax.axis('off')
    ax.set_title(f"Topik {topic_idx}")

plt.show()

def get_top_keywords_only(lda_model, feature_names, n_top_words=6):
    keywords_per_topic = []
    for topic_weights in lda_model.components_:
        top_indices = topic_weights.argsort()[::-1][:n_top_words]
        top_keywords = [feature_names[i] for i in top_indices]
        keywords_per_topic.append(top_keywords)
    return keywords_per_topic

# Contoh penggunaan
top_keywords = get_top_keywords_only(final_lda, feature_names, 6)

# Menampilkan kata kunci tiap topik
for idx, keywords in enumerate(top_keywords):
    print(f"Topik {idx} kata kunci: {keywords}")
```
